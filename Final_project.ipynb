### preprocessing

```python
# covtype, treating the largest class as the positive and the rest as negative
largest = covtype.Cover_Type.value_counts().idxmax() 
new_type = (covtype.Cover_Type == largest)
covtype.drop('Cover_Type',axis = 1, inplace = True) 
covtype['new_type'] = new_type
covtype.head()
```

```python
# letter, uses letters A-M as positives and the rest as negatives
new_lettr = (letter.lettr <= 'M')
letter.drop('lettr',axis = 1, inplace = True) 
letter['new_lettr'] = new_lettr
letter.head()
```

```python
# adult, change y
new_y = (adult.y == ' >50K')
adult.drop('y',axis = 1, inplace = True) 
adult['new_y'] = new_y
```

```python
# perform one hot encoding on adult
adult_nint = adult.select_dtypes(exclude=['int64','bool']).columns
adult = pd.get_dummies(adult, columns=adult_nint)
```
